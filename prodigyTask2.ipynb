```python
import time
import keras_cv
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
model = keras_cv.models.StableDiffusion(img_width=512, img_height=512)
```

```python
images = model.text_to_image("photograph of an astronaut riding a horse", batch_size=5)


def plot_images(images):
    plt.figure(figsize=(20, 20))
    for i in range(len(images)):
        ax = plt.subplot(1, len(images), i + 1)
        plt.imshow(images[i])
        plt.axis("off")


plot_images(images)
```

```python
images = model.text_to_image(
    "A sunset landscape in autumn, with a traveler on a cliff overlooking a forest of orange and red trees. The traveler wears a dark cloak, their silhouette lit by golden sunlight. Soft clouds fill the sky, with rays breaking through. Mist rises from the forest below, giving depth. Use natural colors, warm tones, and soft shadows for a nostalgic atmosphere.",
    batch_size=5
)

plot_images(images)
```
